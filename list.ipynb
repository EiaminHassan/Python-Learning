## Python Lists and List Methods
- Lists are ordered, mutable collections of items in Python. They can contain items of different data types.

```python
my_list = [1, "Hello", 3.14, True]
print(my_list)
```

```python
# Accessing elements
print(my_list[1])  # Output: Hello
```

```python
# Negative indexing
print(my_list[-1])  # Output: True
```

```python
# Range of indices
print(my_list[1:3])  # Output: ['Hello', 3.14]

print(my_list[:4])  # Output: [1, 'Hello', 3.14, True]
```

```python
# Check if Item Exists
if "Hello" in my_list:
    print("Item found!")
```

```python
# Change list items
my_list[0] = 42
print(my_list)  # Output: [42, 'Hello', 3.14, True]
```

```python
# Change a Range of Item Values
list1 = [10, 20, 30, 40, 50]
list1[1:4] = [25, 35, 45]
print(list1)  # Output: [10, 25, 35, 45, 50]

# If you insert less items than you replace, the new items will be inserted where you specified, and the remaining items will move accordingly.
thislist = ["apple", "banana", "cherry"]
thislist[1:3] = ["watermelon"]
print(thislist)
```

```python
# insert items
thislist = ["apple", "banana", "cherry"]
thislist.insert(1, "orange")
print(thislist)  # Output: ['apple', 'orange', 'banana', 'cherry']
```

```python
# Append items
thislist.append("grape")
print(thislist)  # Output: ['apple', 'orange', 'banana', 'cherry', 'grape']
```

```python
# Extend list
thislist.extend(["kiwi", "mango"])
print(thislist)
```

```python
# remove specific item
thislist.remove("banana")
print(thislist)
```

```python
# remove specified index
thislist.pop(1)
print(thislist)  # Output: ['apple', 'banana', 'cherry', 'grape', 'kiwi', 'mango']
```

```python
# Loop through a list
list2 = [10, 20, 30, 40, 50]

for item in list2:
    print(item, end=' ')  # Output: 10 20 30 40 50
    
print()

for index in range(len(list2)):
    print(list2[index], end=' ')  # Output: 10 20 30 40 50
```

```python
# Loopoing using list comprehension
list3 = [1,2,3,4,5]

squared = [x**2 for x in list3]
print(squared)  # Output: [1, 4, 9, 16, 25]
```

```python
# another example of list comprehension

# newlist = [expression for item in iterable if condition == True]


even = [x for x in list3 if x % 2 == 0]
print(even)  # Output: [2, 4]
```

```python
# Sorting lists
unsorted_list = [5, 2, 9, 1, 5, 6]
sorted_list = sorted(unsorted_list)
print(sorted_list)  # Output: [1, 2, 5, 5, 6, 9]

sorted_list_desc = sorted(unsorted_list, reverse=True)
print(sorted_list_desc)  # Output: [9, 6, 5, 5, 2, 1]
```

```python
# custom sort function
# Sort the list based on how close the number is to 50:

def myfunc(n):
  return abs(n - 50)

thislist = [100, 50, 65, 82, 23]
thislist.sort(key = myfunc)
print(thislist)
```
