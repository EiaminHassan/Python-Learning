## Integer & Float numbers in Python with operations

```python
num1 = 10
num2 = 3.5

print(num1)
print(num2)
```

```python
# Arithmetic Operations
addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2
division1 = num1 / num2  # Float division
division2 = num1 // num2  # Integer division

print("Addition:", addition)
print("Subtraction:", subtraction)
print("Multiplication:", multiplication)
print("Float Division:", division1)
print("Integer Division:", division2)
```

```python
# Some python methods for numeric data

print(type(num1))  # Output: <class 'int'>
print(type(num2))  # Output: <class 'float'>

print(isinstance(num1, int))  # Output: True
print(isinstance(num2, float))  # Output: True

print(float(num1))  # Convert int to float
print(int(num2))    # Convert float to int

```

```python
# Exponentiation
exp = num1 ** 2
print("Exponentiation:", exp)
```

```python
# Modulus
modulus = num1 % 3  # 10 % 3 = 1
print("Modulus:", modulus)
```

```python
# Absolute Value
a = -20
absolute_value = abs(-a)
print("Absolute Value:", absolute_value)
```

```python
# round
b = 3.8906
rounded_value = round(b, 2)  # Rounds to 2 decimal places
print("Rounded Value:", rounded_value)
```
