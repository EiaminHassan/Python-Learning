## String Manipulation in Python

```python
my_string = "Hello, World!"
my_list = [1, 2, 3, 4, 5]
```

```python
print(my_string)
print(my_list)
```

String Slicing

```python
print(my_string[0:5])  # Output: Hello
print(my_string[7:])   # Output: World!
```

```python
# positive indexing
print(my_string[0])    # Output: H

# negative indexing
print(my_string[-1])   # Output: !
```

```python
# string reversal

# my_string[start:stop:step]

print(my_string[::-1])  # Output: !dlroW ,olleH
```

```python
# Get the last 3 characters
print(my_string[-3:])  # Output: ld!
```

```python
# Get the length of the string
print(len(my_string))  # Output: 13
```

```python
# convert to lower case characters
print(my_string.lower())  # Output: hello, world!
```

```python
# convert to upper case characters
print(my_string.upper())  # Output: HELLO, WORLD!
```

```python
# count occurrences of a character
temp = "Hello, I'm Eiamin"
print(temp.count('i'))  # Output: 2

# count the occurrences of a substring
print(temp.count('im'))  # Output: 0
print(temp.count("Hello"))  # Output: 1
```

```python
# find index of first occurrence of a character
print(temp.find('E'))  # Output: 11

print(temp.find('z')) # Output: -1 (not found)
```

```python
# replace substring
temp = temp.replace("Eiamin", "Shanto")  # Output: Hello, I'm Shanto
print(temp)
```

```python
# concatenate strings
greeting = "Hello"
name = "Eiamin"

message = greeting + ", " + name
print(message)  # Output: Hello, Eiamin

# another way to concatenate strings
message = f"{greeting}, {name}. Welcome!"
print(message)  # Output: Hello, Eiamin. Welcome!
```

```python
# print a substring with step
print(my_string[0:12:2])  # Output: Hlo ol!
```
