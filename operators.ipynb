## Python Operators and Expressions

```python
# Example of arithmetic operators in Python
print(10 + 5)
```

```python
# Python Walrus Operator (:=), It lets you assign a value to a variable AND use it in the same line.
if (n := len("Hello, World!")) > 10:
    print(f"The length of the string is {n}, which is greater than 10.")
else:
    print(f"The length of the string is {n}, which is not greater than 10.")
```

```python
# Membership operator example
my_list = [1, 2, 3, 4, 5]
if 3 in my_list:
    print("3 is present in the list.")

if 6 not in my_list:
    print("6 is not present in the list.")
```

```python
# Bitwise operator example
a = 5  # In binary: 0101
b = 3  # In binary: 0011
print(f"a & b = {a & b}")  # Bitwise AND
print(f"a | b = {a | b}")  # Bitwise OR
print(f"a ^ b = {a ^ b}")  # Bitwise XOR
print(f"~a = {~a}")        # Bitwise NOT
print(f"a << 1 = {a << 1}")  # Left shift
print(f"a >> 1 = {a >> 1}")  # Right shift
```
