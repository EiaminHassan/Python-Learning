## A tuple is a collection which is ordered and unchangeable. Tuples are written with round brackets.

```python
# Example_1
thistuple = ("apple", "banana", "cherry")
print(thistuple)
```

Tuple items are ordered, unchangeable, and allow duplicate values.
Indexing same as lists.
Tuples are immutable, meaning that you cannot change, add, or remove items after the tuple has been created.

```python
# Tuples allow duplicate values:
thistuple = ("apple", "banana", "cherry", "apple", "cherry")
print(thistuple)
```

```python
# A tuple with strings, integers and boolean values:
tuple1 = ("abc", 34, True, 40, "male")
print(tuple1)
```

```python
# From Python's perspective, tuples are defined as objects with the data type 'tuple':
print(type(tuple1))
```

```python
# We can create a tuple with tuple() constructor:
tuple2 = tuple(("apple", "banana", "cherry")) # note the double round-brackets
print(tuple2)
```

Accessing, Ranging, Slicing etc. same as lists.

Once a tuple is created, you cannot change its values. Tuples are unchangeable, or immutable as it also is called.

But there is a workaround. You can convert the tuple into a list, change the list, and convert the list back into a tuple.

```python
#Example_2
x = ("apple", "banana", "cherry")
y = list(x)
y[1] = "kiwi"
x = tuple(y)

print(x)
print(type(x))
```
