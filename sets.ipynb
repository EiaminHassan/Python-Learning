## A set is a collection which is unordered, unchangeable*, and unindexed.

Set don't allow duplicate values.

```python
thisset = {"apple", "apple", "banana", "cherry"}
print(thisset) #Sets are unordered, so you cannot be sure in which order the items will appear.
```

## Indexing are not supported in sets.

```python
print(thisset[1]) # Give you an error
```

```python
# Add set items
thisset.add("orange")
print(thisset)
```

```python
# Add any iterable (sets, tuples, lists, dictionaries etc.)
thisset.update({"mango", "grapes", "orange"})
print(thisset)
print(type(thisset))

thisset.update([1, 2, 3])
print(thisset)
```

## Python - Remove Set Items

```python
# Remove "banana" by using the remove() method:
thisset = {"apple", "banana", "cherry"}
thisset.remove("banana")
print(thisset)

# Remove "banana" by using the discard() method:
thisset = {"apple", "banana", "cherry"}
thisset.discard("banana")
print(thisset)
```

```python
# The clear() method empties the set:

thisset = {"apple", "banana", "cherry"}
thisset.clear()
print(thisset)
```

## Join sets

```python
# The union() and update() methods joins all items from both sets.
set1 = {"a", "b", "c"}
set2 = {1, 2, 3}

set3 = set1.union(set2)
print(set3)

# or
set1 = {"a", "b", "c"}
set2 = {1, 2, 3}

set3 = set1 | set2 # another way to do union
print(set3)


set1 = {"a", "b", "c"}
set2 = {1, 2, 3}
set3 = {"John", "Elena"}
set4 = {"apple", "bananas", "cherry"}

myset = set1.union(set2, set3, set4) # join multiple sets
print(myset)

# or
set1 = {"a", "b", "c"}
set2 = {1, 2, 3}
set3 = {"John", "Elena"}
set4 = {"apple", "bananas", "cherry"}

myset = set1 | set2 | set3 |set4
print(myset)
```

```python
# Intersection
# Keep ONLY the duplicates

set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1.intersection(set2)
print(set3)

# or
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1 & set2
print(set3)
```

```python
# Difference
# The difference() method will return a new set that will contain only the items from the first set that are not present in the other set.

set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1.difference(set2)

print(set3)

# or 
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1 - set2
print(set3)
```

```python
# The symmetric_difference() method will keep only the elements that are NOT present in both sets.
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1.symmetric_difference(set2)

print(set3)

# or
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1 ^ set2
print(set3)
```
